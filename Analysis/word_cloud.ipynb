```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf"
WordCloud(font_path=None)  # works if default font supports your characters
 

file_path = "/home/avishka/Cloud/Project/MapReduce/output.txt"

positive_words = {}
negative_words = {}

with open(file_path, "r") as file:
    for line in file:
        parts = line.strip().split('\t')
        if len(parts) == 3:
            sentiment, word, score = parts
            try:
                score = float(score)
                word = word.strip()
                if word:  # Ensure word is not empty
                    if sentiment == "positive":
                        positive_words[word] = score
                    elif sentiment == "negative":
                        negative_words[word] = score
            except ValueError:
                continue

# Print word count to debug
print(f"Loaded {len(positive_words)} positive and {len(negative_words)} negative words.")

# Generate word clouds

positive_wc = WordCloud(
    width=600,
    height=400,
    background_color='white',
    colormap='Greens',
    font_path=font_path
)

negative_wc = WordCloud(
    width=600,
    height=400,
    background_color='white',
    colormap='Reds',
    font_path=font_path
)



positive_wc.generate_from_frequencies(positive_words)
negative_wc.generate_from_frequencies(negative_words)

# Plot
fig, axes = plt.subplots(1, 2, figsize=(14, 7))
axes[0].imshow(positive_wc, interpolation='bilinear')
axes[0].set_title('Positive Words (Avg Rating)', fontsize=16)
axes[0].axis('off')

axes[1].imshow(negative_wc, interpolation='bilinear')
axes[1].set_title('Negative Words (Avg Rating)', fontsize=16)
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
