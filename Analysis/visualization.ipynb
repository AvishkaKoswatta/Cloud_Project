```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data using the correct column names
df = pd.read_csv("/home/avishka/Cloud/Project/MapReduce/output.txt", sep="\t", names=["category", "word", "rating"])

# Separate into negative and positive
neg_df = df[df["category"] == "negative"].sort_values(by="rating")
pos_df = df[df["category"] == "positive"].sort_values(by="rating", ascending=False)

# Set plot style
sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Plot negative words
sns.barplot(x='rating', y='word', data=neg_df, palette='Reds_r', ax=axes[0])
axes[0].set_title('Negative Words by Average Rating')
axes[0].set_xlabel('Rating')
axes[0].set_ylabel('Word')

# Plot positive words
sns.barplot(x='rating', y='word', data=pos_df, palette='Greens', ax=axes[1])
axes[1].set_title('Positive Words by Average Rating')
axes[1].set_xlabel('Rating')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```
